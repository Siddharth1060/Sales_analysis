### Combining dfs into 1 df

```python
df= pd.DataFrame()
```

```python
df=pd.concat([df_jan,df_feb,df_mar,df_apr,df_may,df_jun,df_jul,df_aug,df_sep,df_oct,df_nov,df_dec])
```

```python
df.reset_index(inplace= True)
```

```python
df.drop(columns =['index'], inplace=True)
```

```python
df.loc[df.Product.isna()==True]
```

#### Data Cleaning & Feature engineering 

```python
df.info()
```

#### Order Date: converting to datetime and splitting into 2 columns: Date and Time

```python
df['Date']=df['Order Date'].str.strip().str[:-6]
```

```python
df.Date= pd.to_datetime(df['Date'], errors='coerce')
```

```python
df.loc[df.Date.isna()==True]
```

```python
df.iloc[1065:1075,]
```

```python
#dropping all Nans from the dataframe, including the headings encountered, eg row 1073
```

```python
df.dropna(inplace=True)
```

```python
df['Time'] =df['Order Date'].str.strip().str[-5:]
```

```python
df['Hour']= df.Time.str.split(':').str[0]
```

```python
df['Minute']= df.Time.str.split(':').str[1]
```

```python
pd.to_datetime(df['Time'], format='%H:%M')
```

```python
#dropping 'Order Date' column
```

```python
df.drop(columns=['Order Date'], inplace=True)
```

```python
df
```

#### Purchase Address: Getting city  and State

```python
df['City'] = df['Purchase Address'].str.split(',').str[1]
```

```python
df['State'] = df['Purchase Address'].str.split().str[-2]
```

#### Quantity Ordered, Price Each: converting to numeric

```python
df['Quantity Ordered']= pd.to_numeric(df['Quantity Ordered'])
```

```python
df['Price Each']= pd.to_numeric(df['Price Each'])
```

#### What was the best month for sales? What was the earning that month?

```python
df['Order_Price']=df['Price Each']*df['Quantity Ordered']
```

```python
sale_by_mnth=df.groupby(df.Date.dt.month).sum().sort_values('Order_Price', ascending= False).reset_index()
```

```python
sale_by_mnth
```

```python
plt.figure(figsize=(10,5))
plt.bar(sale_by_mnth.Date,sale_by_mnth.Order_Price)
plt.xticks(sale_by_mnth.Date)
plt.xlabel('Months of year 2019')
plt.ylabel('Sales by Month(in Million $)')
```

#### Which city had the highest number of sales?

```python
df
```

```python
sale_by_city =df.groupby([df.City,df.State]).sum().sort_values('Order_Price', ascending= False)
```

```python
sale_by_city.reset_index(inplace=True)
```

```python
sale_by_city.City=sale_by_city.City+'('+ sale_by_city.State+')'
```

```python
sale_by_city
```

```python
plt.figure(figsize=(17,5))
plt.bar(sale_by_city.City,sale_by_city.Order_Price)
plt.ylabel('Sales by City(in Million $)')
```

#### What time do we show ads for customers to buy our products?

```python
df
```

```python
 order_cnt_by_time= df.groupby(['Hour']).count().sort_values('Hour', ascending=True).iloc[:,0:1]
```

```python
order_cnt_by_time.reset_index(inplace=True)
```

```python
order_cnt_by_time.rename(columns={'Order ID': 'Order_count'},inplace=True)
```

```python
order_cnt_by_time
```

```python
plt.figure(figsize=(10,5))
plt.plot(order_cnt_by_time.Hour,order_cnt_by_time.Order_count)
plt.xlabel('Time of Order(00=12 a.m., 23=> 11 p.m.)')
plt.ylabel('Count of orders by Time (in 100,000$)')
```
